#Mane 3x3 size er 20 ta filter ei input image er jnno kaj korbe..
#Filter basically edge, boundary esob detect kore and for better detection amra 20ta filter dicchi.


```python
model=Sequential()
model.add(Conv2D(input_shape=(10,10,3),
                filters=10,kernel_size=(3,3), #--->[3*3] kernel matrix & strides (1,1) mens --> 1 ghor kore barbe.
                strides=(1,1),padding='valid'))
```

```python

print(model.output_shape)
```

```python

model=Sequential()
model.add(Conv2D(input_shape=(10,10,3), #[Height,wight, and color(RGB=3)]
                filters=10,kernel_size=(6,6),  # filters=10----> mne (3*3) size er 10 ta filter ei input images er jnne kaj korbe,filter mainly edge boundary.
                strides=(1,1),padding='same'))
```

```python
print(model.output_shape)
```

```python
model.add(MaxPool2D(pool_size=(2,2),padding='same'))  #(pool_size = (2, 2) prothome 1ghor jbe pore abe 1 ghor jbe kintu row 1 ta and columns 1 ta kore ibe mne (2,2)
```
